```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('Titanic_data.csv')
```

```python
titanic_df.head()
```

```python
Pie Chart DONE
Bar Graph DONE
Histogram DONE
Scratter Plot (x-y) DONE
Box Plot
Heatmap
```

```python
# Bargraph
titanic_df['Survived'].value_counts().plot(
    kind ='barh', title = "Bar Graph of Passenger survived rate",
    legend = True, grid= True
)
```

```python
import matplotlib.pyplot as plt
x = titanic_df['Survived'].value_counts()
plt.bar(['Survived','Dead'], x.to_numpy())
plt.show()
plt.close()
```

```python
# Pie Chart

titanic_df['Embarked'].value_counts().plot.pie(
    title = "Pie Chart", legend = True, grid = True
)
```

```python
# Histogram
titanic_df['Age'].plot.hist(bins =20,colormap = 'Set1')
```

```python
# Scratter Plot (x-y)
titanic_df.plot.scatter('Age','Fare')
```

```python
# Box Plot
titanic_df.boxplot(
    column=['Age']
)
```
