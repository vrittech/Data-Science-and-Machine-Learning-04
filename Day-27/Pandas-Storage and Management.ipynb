```python
import pandas as pd
import numpy as np
```

```python
titanic_df = pd.read_csv('Titanic_data.csv')
titanic_df.head()
```

```python
titanic_df.info()
```

### dtype()

```python
titanic_df.dtypes
```

```python
titanic_df['Age'].fillna(42, inplace = True)
titanic_df['Age'] = titanic_df['Age'].astype(np.int32)
```

```python
titanic_df.dtypes
```

```python
titanic_df = titanic_df.astype(
    {"Survived":np.float64,"Fare":np.int64}
)
titanic_df.dtypes
```

### sample()

```python
titanic_df.sample(n=13, random_state=99)
```

```python
titanic_df.sample(frac=0.015, ignore_index=True, random_state=99)
```

### index()

```python
titanic_df
```

```python
# set_index()
titanic_df.set_index('PassengerId', inplace=True)
titanic_df.head()
```

```python
# reset_index()
titanic_df = titanic_df.dropna(axis=0)
titanic_df.head(10)
```

```python
titanic_df.reset_index(drop=True, inplace = True)
```

### rename()

```python
titanic_df.rename({"Pclass":'Ptype',"Sex":'Gender'}, axis = 1, inplace = True)
```
