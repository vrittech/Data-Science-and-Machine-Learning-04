```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import pickle

LABELS = {
    'Iris-setosa':0,
    'Iris-versicolor':1, 
    'Iris-virginica':2
}
```

```python
dataset = pd.read_csv("IRIS.csv")
dataset.head()
```

```python
features = dataset.drop('species', axis=1).to_numpy()
target = dataset['species'].replace(LABELS).to_numpy()

# Standard Scaler
scaler = StandardScaler()
scaled_feature = scaler.fit_transform(features)

# SPlit to train test
X_train, X_test, y_train, y_test = train_test_split(scaled_feature, target, test_size=0.2, stratify=target, shuffle=True, random_state=88)
```

```python
# Training 
logistic_reg = LogisticRegression(class_weight='balanced')
logistic_reg.fit(X_train,y_train)
```

```python
# Evaluation
pred = logistic_reg.predict(X_test)
print(classification_report(y_test, pred))
```

```python
# Model saving
#1. Scaler
with open("best_model/standard_scaler.pkl","wb") as file:
    pickle.dump(scaler, file)
#2. Model
with open("best_model/reg_logistic_model.pkl","wb") as file:
    pickle.dump(logistic_reg, file)
```

## Prediction Pipeline

```python
import pickle
import numpy as np

LABELS = {
    0: 'Iris-setosa',
    1 : 'Iris-versicolor', 
    2 : 'Iris-virginica'
}

user_input = np.array([[5.55, 3.01, 1.9, 0.77]])
```

```python
def get_save_file(file_path:str):
    
    with open(file_path,"rb") as file:
        output = pickle.load(file)
        
    return output
```

```python
best_scaler = get_save_file("best_model/standard_scaler.pkl")
best_model = get_save_file("best_model/reg_logistic_model.pkl")
```

```python
model_output = best_model.predict(
    best_scaler.transform(
        user_input
    )
)
```

```python
user_ouput = LABELS[model_output[0]]
print(user_ouput)
```
