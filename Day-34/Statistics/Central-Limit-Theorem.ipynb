## Central Limit Theorem


The Central Limit Theorem is a fundamental concept in statistics that describes the behavior of the sum or average of a large number of independent and identically distributed (i.i.d.) random variables. It states that regardless of the shape of the original population distribution, the sampling distribution of the sum or average tends to follow a normal distribution as the sample size increases.

The CLT has immense practical significance since it allows us to make statistical inferences about population parameters even when we don't know the underlying distribution. It forms the basis for many statistical techniques and hypothesis tests.


### Assumptions of the Central Limit Theorem
Before we dive into the details, let's outline the assumptions of the Central Limit Theorem:

- **Independence:** The random variables being sampled must be independent of each other.
- **Identical Distribution:** The random variables must have the same distribution, with the same mean and standard deviation.
- **Sample Size:** The sample size should be sufficiently large. The CLT usually works well for sample sizes greater than 30, although there are cases where it can still provide reasonably accurate results for smaller sample sizes.

### Statement of the Central Limit Theorem
The Central Limit Theorem can be stated as follows:

Let $X_1, X_2, \ldots, X_n$ be a sequence of $n$ **independent and identically distributed** random variables with mean $\mu$ and standard deviation $\sigma$. Consider the random variable $Y_n = \frac{X_1 + X_2 + \ldots + X_n}{\sqrt{n}}$. Then, as $n$ approaches infinity, the distribution of $Y_n$ approaches a standard normal distribution with mean $0$ and standard deviation $1$.


In simpler terms, the sum or average of a large number of independent and identically distributed random variables follows an approximately normal distribution.


## Demonstrating CLT

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Uniform Distribution

```python
np.random.seed(99)

population = np.random.uniform(0, 1, size = 100000)

plt.figure(figsize=(4,3))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.8)
plt.title("Population Distribution (Uniform)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Sample size for experiment
sample_size = [5,10,30,100,500,1000]

# Generate sample of diff size and calculate sample means
sample_means = []
for n in sample_size:
    samples = np.random.choice(population, size=(1000,n), replace = True)
    means = np.mean(samples, axis =1 )
    sample_means.append(means)

# Plot the distribution of samples means
plt.figure(figsize=(10,6))
for i, means in enumerate(sample_means):
    plt.subplot(2 ,3 , i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.4)
    plt.title(f"Sample Size :{sample_size[i]}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
plt.tight_layout()
plt.show()
```
